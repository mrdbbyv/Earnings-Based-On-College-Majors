# Visualizing Earnings Based On College Majors¶

This project will be focused on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

We will try to answer questions below using visualitions

- Do students in more popular majors make more money?

- How many majors are predominantly male? Predominantly female?

- Which category of majors have the most students?

## Relationships exploration via Scatter Matrix Plots

```python
# This module needs to be imported separately
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

There is no relationship between  `Median` and `Sample_size`.

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

Again there is not any relationship between `Median`, `Sample_size` and `Unemployment_rate`.

## Comparison using Bar Plots

Using bar plot to compare the percentages of `ShareWomen` from first ten and last ten rows of the `recent_grads` dataframe.

```python
recent_grads[:10]['ShareWomen'].plot(kind='bar')
```

```python
recent_grads[-10:]['ShareWomen'].plot(kind='bar')
```

The last ten majors have a greater share of women than the first ten majors.
